```python
#biểu đồ nhiều đương
#Cho dữ liệu sau
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0,10,100)
y1=np.sin(t)
y2=np.cos(t)
y3=np.sin(t) * np.cos(t)

#tạo biểu đồ các đường theo thời gian
plt.figure(figsize=(8, 5))
plt.plot(t, y1, label='sin(t)', color='blue')
plt.plot(t, y2, label='cos(t)', color='red')
plt.plot(t, y3, label='sin(t)*cos(t)', color='green')
plt.title("biểu đồ")
plt.xlabel("thời gian t")
plt.ylabel("giá trị")
plt.legend()
plt.grid(True)







```

```python
#Cho dữ liệu sau
labels = ["Q1","Q2","Q3","Q4"]
A = [500,700,800,600]
B = [450,350,650,750]
C = [500,250,850,600]

##########Vẽ biểu đồ thanh nhòm thể hiện doanh thu của A,B,C trong 4 quý
# Xác định vị trí cho các cột
x = np.arange(len(labels))
width = 0.25  # Độ rộng của cột

# Vẽ biểu đồ thanh nhóm
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(x - width, A, width, label='A')
ax.bar(x, B, width, label='B')
ax.bar(x + width, C, width, label='C')

# Thiết lập nhãn và tiêu đề
ax.set_xlabel('Quý')
ax.set_ylabel('Doanh thu')
ax.set_title('Doanh thu của A, B, C trong 4 quý')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```
